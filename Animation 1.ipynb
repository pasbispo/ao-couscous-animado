```python
h=6.62607004e-34
hc = h/(2*np.pi)
c=299.792458 
e2 = (1.6e-19)**2
e4 = e2*e2
eps0 = 8.85e-12
m = 9.11e-31
a_0 = 4*np.pi*eps0*hc*hc/(m*e2)
beta = e4/(12*np.pi*eps0**2*c**3*m**2)

T = np.linspace(0,1,1000)
dt = T[1]-T[0]
L = 5*beta*m/a_0
```
